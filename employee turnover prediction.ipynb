```python
total = data.groupby('Department')['Attrition'].count().reset_index().sort_values('Attrition', ascending=False)
p1=sns.barplot(data=total, x='Department', y='Attrition')
p1.bar_label(p1.containers[0])
plt.title('Top departments of attritions')
plt.xticks(rotation=0)
plt.show()
```

```python
data.groupby('JobRole')['Attrition'].count()
```

```python
total = data.groupby('JobRole')['Attrition'].count().reset_index().sort_values('Attrition', ascending=False)
p1=sns.barplot(data=total, x='JobRole', y='Attrition')
p1.bar_label(p1.containers[0])
plt.title('Top job roles of attritions')
plt.xticks(rotation=90)
plt.show()
```
